```python
def generationofparanthesis(n):
    stack=[]
    res=[]
    def backtrack(openp,close):
        if openp == close == n:
            res.append("".join(stack))
            return
        if openp<n:
            stack.append("(")
            backtrack(openp+1,close)
            stack.pop()
        if close<openp:
            stack.append(")")
            backtrack(openp,close+1)
            stack.pop()
    backtrack(0,0)
    return res
print(generationofparanthesis())


    
        
```
